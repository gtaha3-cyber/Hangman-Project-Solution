## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to guess the fruit name correctly.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible in the output.
> ##### In this way, all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, the number of chances equals the word length + 2.
> ##### For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.
> ##### Make game repeats until player ends the game.
> ##### Bonus : Create a function that takes 3 parameters `Easy`, `Medium` and `Hard` where `Medium` has more difficult words to be guessed than `Easy` and so on.

```python
import random
import os
import shutil


list_easy = ['mango', 'apple', 'orange']
list_medium = ['banana', 'cherry', 'avocado']
list_hard = ['strawberry', 'watermelon', 'blueberry']


def choose_word(list_easy: list, list_medium: list, list_hard: list, user_level_selection: str) -> str:
    """
    Pick a random secret word based on the chosen difficulty level.

    Args:
        list_easy (list): List of easy words.
        list_medium (list): List of medium words.
        list_hard (list): List of hard words.
        user_level_selection (str): The difficulty chosen by the user (already upper-cased).

    Returns:
        str: A random word from the list matching the given level (defaults to Easy if invalid input).
    """
    if user_level_selection == 'EASY':
        word_list = list_easy

    elif user_level_selection == 'MEDIUM':
        word_list = list_medium

    elif user_level_selection == 'HARD':
        word_list = list_hard

    else:
        word_list = list_easy

    return random.choice(word_list)


def create_display(word: str) -> list:
    """
    Create the initial display list made of underscores, one for each letter.

    Args:
        word (str): The secret word.

    Returns:
        list: A list of underscores with the same length as the word.
    """
    return ['_' for letter in word]


def update_display(word: str, guess: str, display: list) -> list:
    """
    Reveal the guessed letter in the display list if it exists in the word.

    Args:
        word (str): The secret word.
        guess (str): The letter guessed by the user (already lower-cased).
        display (list): The current display list (underscores/revealed letters).

    Returns:
        list: The updated display list.
    """
    for index in range(len(word)):
        if word[index] == guess:
            display[index] = guess

    return display


while True:

    user_level_selection = input('Enter your difficulty level (Easy / Medium / Hard)').upper()

    secret_word = choose_word(list_easy, list_medium, list_hard, user_level_selection)

    chances = len(secret_word) + 2

    display = create_display(secret_word)

    print('Word: ' + ' '.join(display))

    for i in range(chances):

        guess = input('Guess a letter: ').lower()

        if guess in secret_word:
            display = update_display(secret_word, guess, display)

            print('Word: ' + ' '.join(display))

            if '_' not in display:
                print('Congratulations! You guessed the word: ' + secret_word)
                break

        else:
            print('Wrong Guess!')

    else:
        print('Game Over, Better Luck Next Time! The word was: ' + secret_word)

    play_again = input('Would you like to Play Again ? [Y/N]').upper()

    if play_again == 'N':
        print('\nThank you for playing Hangman Game :D')
        break
```
